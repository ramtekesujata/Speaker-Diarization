Step 1. Pre-Emphasis Filter

```python
def emphasis_signal(signal):
    pre_emphasis = 0.97
    return np.append(signal[0], signal[1:] - pre_emphasis * signal[:-1])
```

```python
def audio_emphasised(audio_data):
  audio_data = emphasis_signal(audio_data)
  return audio_data
```

Step 5: Feature Extraction

Calculate Mel- Frequency Cepstral Coefficients (MFCC) features, their derivatives and double derivatives.

```python
def get_mfcc_feature(wave, sr):
    mfccs = librosa.feature.mfcc(y= wave, sr=sr,n_mfcc=13)
    # Creating derivatives
    delta_mfccs = python_speech_features.base.delta(mfccs,1)
    delta_delta_mfccs = python_speech_features.base.delta(delta_mfccs,1)
    # Concatenating equals appending
    mfccs = np.concatenate((mfccs,delta_mfccs))
    mfccs = np.concatenate((mfccs,delta_delta_mfccs))
    return mfccs
```

```python
def extract_feature(frames,sr):
    all_features = [] # similar to vector<vector<int>>
    for arr in tqdm(frames):
        features_list = []  # similar to vector<int>
        mfccs = get_mfcc_feature(arr, sr)
        # Adding additional features from librosa
        features_list.append(np.mean(librosa.feature.chroma_stft(y=arr, sr=sr)))
        features_list.append(np.mean(librosa.feature.rms(y=arr)))
        features_list.append(np.mean(librosa.feature.spectral_centroid(y=arr, sr=sr)))
        features_list.append(np.mean(librosa.feature.spectral_bandwidth(y=arr, sr=sr)))
        features_list.append(np.mean(librosa.feature.spectral_rolloff(y=arr, sr=sr)))
        features_list.append(np.mean(librosa.feature.zero_crossing_rate(arr)))
        # Adding features from MFCC
        for mfcc in mfccs:
            features_list.append(np.mean(mfcc))
        all_features.append(features_list)
    return all_features
```

```python
features = extract_feature(frames,8000)
print(features[0])
```

Step 7: Spectral Clustering

```python
from spectralcluster import RefinementOptions
rf= RefinementOptions(gaussian_blur_sigma=1,p_percentile=0.95)
```

```python
clusterer = SpectralClusterer(
    min_clusters=n_clusters,
    max_clusters=n_clusters+5,
    autotune=None,
    laplacian_type=None,
    custom_dist="cosine",
    refinement_options=rf)

labels = clusterer.predict(df_scaled.values)
```
